```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head(20)
```

```python
# How many neighborhood groups are available and which shows up the most?

hoods = air_bnb.drop_duplicates('neighbourhood_group').count()[['neighbourhood_group']]
hoods
```

```python
# big_hoods = air_bnb.count()[['neighbourhood_group']]
# big_hoods

big_hoods = air_bnb.groupby('neighbourhood_group').count()
big_hoods.sort_values('id', ascending=False)

# There are 5 and Manhattan shows up the most with a count of 21661
```

```python
# Are private rooms the most popular in manhattan?

private = air_bnb.groupby(['neighbourhood_group', 'room_type']).count()[['number_of_reviews']]
private.sort_values('number_of_reviews', ascending=False).head(40)

# No, entire home/apt are the most popular in Manhattan. Private rooms are second.
```

```python
# Which hosts are the busiest and based on their reviews?

hosts = air_bnb.groupby('host_name').count()[['number_of_reviews']]
hosts.sort_values('number_of_reviews', ascending=False).head(10)

#  Michael is the busiest based on his 417 reviews
```

```python
#Which neighorhood group has the highest average price?
pricey = air_bnb.groupby('neighbourhood_group').mean()[['price']]
pricey.sort_values('price', ascending=False).head()

#  Manhattan
```

```python
# Which neighbor hood group has the highest total price?

bigmoney = air_bnb.groupby('neighbourhood_group').sum()[['price']]
bigmoney.sort_values('price', ascending=False).head()


# Manhattan
```

```python
#Which top 5 hosts have the highest total price?

ny = air_bnb.groupby('host_name').sum()[['price']]
ny.sort_values('price', ascending=False).head()

# Sounder(NYC)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

# avail = air_bnb[air_bnb['availability_365'] == 0].groupby('host_name')
# x = avail[avail['number_of_reviews' >= 100]
# x.head(30)
    
avail = air_bnb.loc[((air_bnb['availability_365'] == 0) & (air_bnb['number_of_reviews'] >= 100))].count()
avail

# 162 listing have 0 availability and with a review count of 100 or more.
```

```python
# What host has the highest total of prices and where are they located?

nyc = air_bnb.groupby(['host_name','neighbourhood_group', 'neighbourhood']).sum()[['price']]
nyc.sort_values('price', ascending=False).head()

# Sonder(NYC) in the Finacial District in Manhattan
```

```python
# When did Danielle from Queens last receive a review?
# bhood = air_bnb.set_index('neighbourhood_group')
# bhood.loc['neighbourhood_group', 'host_name']

bhood = air_bnb.loc[air_bnb['host_name'] == 'Danielle']
bhood.sort_values('last_review', ascending=False)

# Last review 2019-07-08
```
